# STATISTICAL EVIDENCE

#Do charges of people who smoke differ significantly from the people who don't

```python
#Try to get 2 different samples for smokers/No smokers for charges 
#Null hypothesis: the two groups have the same mean which means same charges for both smokers and non smokers
#Alternative hypthesis will be that 2 groups have different mean 
#independent groups:
# confident interval taking it as 95% alpha = 0.05

Data.head()
Smoker = Data[Data['smoker'] == 'yes'].charges
NoSmoker = Data[Data['smoker'] == 'no'].charges

Smoker.head()
```

```python
#assuming the mean is same 
# test will be indepedent 2 sample t-test 
import numpy as np
from scipy.stats import ttest_ind
from statsmodels.stats.power import ttest_power
```

```python
t_statistic, p_value = ttest_ind(Smoker, NoSmoker)
print(t_statistic, p_value)
```

```python
if p_value < 0.05:
    print("Since the P values is less than the alpha, we reject the null")
else:
    print("Since the P value is more than alpha, Fail to reject the null")
```

# as Null is rejected we can say theere is difference in charges of smoker and non smokers. 

```python
# SOLVE using the manual way by formula 
# equal sample size and assume equal population variance
t_critical = 1.962
N1 = len(Smoker)
N2 = len(NoSmoker)
d1 = N1-1
d2 = N2-1
df = d1+d2
s1 = np.std(Smoker,ddof=1)
s2 = np.std(NoSmoker,ddof=1)
x1_bar = np.mean(Smoker)
x2_bar = np.mean(NoSmoker)
df
sp = np.sqrt((d1*s1**2 + d2*s2**2)/df)
se = sp*np.sqrt(1/N1 + 1/N2)
t = (x2_bar - x1_bar)/(se)
print("t-statistic", t)
df
```

```python
if t < t_critical:
    print("Since the test statistic is less than the critical value, there is a difference")
else:
    print("There is no change in mean")
```

#Does bmi of males differ significantly from that of females? 

```python
#indepedent 2 Sample T test considering mean is same
# Null will be mean will be same
# alternate Hypothesis will be that mean will be different 
# take 2 df for male and Female with their bmi

Data_m = Data[Data['sex'] == 'male'].bmi
Data_f = Data[Data['sex'] == 'female'].bmi

Data_f.mean(),Data_m.mean()
```

```python
t_statistic, p_value = ttest_ind(Data_m, Data_f)
print(t_statistic, p_value)
```

```python
if p_value < 0.05:
    print("Since the P values is less than the alpha, we reject the null")
else:
    print("Since the P value is more than alpha, Fail to reject the null")
```

#here the P value is more than the alpha we can say that the  sample mean is the same for male and female.

#Is the distribution of bmi across women with no children, one child and two children, the same? 

```python
#one way ANOVA test

Data1 = Data[Data['sex'] == 'female']
zeroc = Data1[Data1['children'] == 0].bmi
onec = Data1[Data1['children'] == 1].bmi
twoc = Data1[Data1['children'] == 2].bmi
```

```python
zeroc.value_counts().sum()
```

```python
onec.value_counts().sum()
```

```python
twoc.value_counts().sum()
```

```python
mean_Bmi = zeroc.mean()
std_Bmi = zeroc.std()
print('Count, Mean and standard deviation of BMI of women with zeroc: %3d, %3.2f and %3.2f' % (len(zeroc), zeroc.mean() ,np.std(zeroc )))
print('Count, Mean and standard deviation of BMI of women with onec: %3d, %3.2f and %3.2f' % (len(onec), onec.mean() ,np.std(onec )))
print('Count, Mean and standard deviation of BMI of women with twoc: %3d, %3.2f and %3.2f' % (len(twoc), twoc.mean() ,np.std(twoc )))

```

```python

mean_Bmi_df = pd.DataFrame()
df1            = pd.DataFrame({'children': '0', 'Mean_Bmi':zeroc})
df2            = pd.DataFrame({'children': '1', 'Mean_Bmi':onec})
df3            = pd.DataFrame({'children': '2', 'Mean_Bmi':twoc})

mean_Bmi_df = mean_Bmi_df.append(df1) 
mean_Bmi_df = mean_Bmi_df.append(df2) 
mean_Bmi_df= mean_Bmi_df.append(df3) 

mean_Bmi_df.head()

```

```python
sns.boxplot(x = "children", y = "Mean_Bmi", data = mean_Bmi_df)
plt.title('Mean BMI for Women based on kids')
plt.show()
```

```python
#𝐻0: 𝜇1 = 𝜇2 = 𝜇3
#𝐻𝐴 : At least one 𝜇 differs 
#we select α = 0.05
#ANOVA uses F-tests to statistically test the equality of means.

import statsmodels.api         as     sm
from   statsmodels.formula.api import ols
 
mod = ols('Mean_Bmi ~ children', data = mean_Bmi_df).fit()
aov_table = sm.stats.anova_lm(mod, typ=2)
print(aov_table)
```

```python
# here the p value ( 0.71) > alpha of 0.05 we fail to reject the Null , So the mean of the BMI across women 
#with different number of kids is same 
```
